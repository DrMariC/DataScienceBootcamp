```python
# Import pandas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame with balance.txt
filename = "balance.txt"
df = pd.read_csv(filename, sep=' ')
df
```

```python
# Comparing the average income based on ethnicity indicates that African Americans earn the most.
df.Ethnicity.describe()
df.Ethnicity
print("The average income of Caucasian's is", df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean())
print("The average income of Asian's is", df[df.Ethnicity == "Asian"].loc[:,"Income"].mean())
print("The average income of African American's is", df[df.Ethnicity == "African American"].loc[:,"Income"].mean())
# Use data viz to understand the data better
plt.title('Bar Graph of Income by Ethnicity')
# Create a list of colors
colors = ['b', 'g', 'r', 'c', 'm']
sns.barplot(data = df, x='Ethnicity', y ='Income', palette = colors)
```

```python
# UPDATE FOR RESUBMISION:
# Instead of looking at the peaks of the datasets, this time I looked at the distribution (Thanks to the reviewer for the hint! ;)
# According to the KDE graps below, one can see how most of the data is distributed around a higher balance on average, for unmerried, vs merried.
# Unmeried distribution is also narrower than that of the married, indicating less spread. 
# According to the violin plot as well as the mean data you printed, on average, unmarried people actually have a slightly higher balance.

# Make a Kernel Density Plots (KDEs)
plt.title('Distribution Graph of Balance by Merital Status')
sns.kdeplot(data = df, x='Balance', hue='Married')
```

```python
# UPDATE FOR RESUBMISION: On average, unmarried people have a slightly higher balance than single people;
# This is clear when comparing the mean values printed (13.493509 > 13.388473),
# as well as the median (another measure of the average) on the violine graph (white line inside a box).
# This could be explained by the fact that having a family reqires more resources, such as bigger housing, more food, etc.
# Therefore there is more money spend and less left on balance.
print(df.groupby(['Married']).mean(numeric_only=True))
# Use data viz to understand the data better
plt.title('Violin Graph of Balance by Merital Status')
# Create a list of colors
colors = ['b', 'g', 'r', 'c', 'm']
sns.violinplot(data = df, x='Married', y ='Balance', palette = colors)
```
